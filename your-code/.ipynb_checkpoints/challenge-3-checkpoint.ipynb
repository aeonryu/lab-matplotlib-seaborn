#### Use the appropriate visulalization to show the distribution of the column `Age`

```python
# your code here
print(titanic['Age'].min())
print(titanic['Age'].max())
plt.hist(titanic['Age'], bins=16) #decidí 16 bins poruqe así se distribuye de 5 en 5 años cada bin
plt.title('Distribución de edades')
plt.xlabel('Edad')
plt.ylabel('Frecuencia')
```

#### Use subplots to plot the distribution of the `Age` variable with bins equal to `10`, `20` and `50`.

```python
# your code here
#sns.set(style="white", palette="muted", color_codes=True) #Funciona pero no es necesario
sns.set()
f, axes = plt.subplots(2, 3, figsize=(7, 7))  #, sharex=True parece ser que es compartir el eje x
sns.despine(left=True)

sns.distplot(titanic['Age'], bins=10, color="m", ax=axes[0, 0])
sns.distplot(titanic['Age'], bins=20, color="m", ax=axes[0, 1])
sns.distplot(titanic['Age'], bins=50, color="m", ax=axes[0, 2])

plt.setp(axes, yticks=[])
plt.tight_layout()
```
